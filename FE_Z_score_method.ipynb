# **trimming**

```python
new_df = df[(df['cgpa']>5.11) & (df['cgpa']<8.80)]
new_df
```

# **Z-score**

```python
df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
```

```python
df.head()
```

```python
df[(df['cgpa_zscore']>3) | (df['cgpa_zscore']<-3)]
```

# **Capping**

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
df['cgpa'] = np.where (
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
)
    )
```
