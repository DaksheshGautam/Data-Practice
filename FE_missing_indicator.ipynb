```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator, SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = sns.load_dataset('titanic')
```

```python
df = df[['age','fare','survived']]
```

```python
df.head()
```

```python
X = df.drop(columns=['survived'])
y = df['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 2)
```

```python
si = SimpleImputer()
```

```python
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
clf = LogisticRegression()
clf.fit(X_train_trf, y_train)
ypred = clf.predict(X_test_trf)
accuracy_score(y_test, ypred)
```

```python
mi = MissingIndicator()
mi.fit(X_train)
X_train_missing = mi.transform(X_train)
X_test_missing = mi.transform(X_test)
```

```python
X_train_missing
```

```python
X_train['age_na'] = X_train_missing
X_test['age_na'] = X_test_missing
```

```python
X_train
```

```python
X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
clf.fit(X_train_trf2, y_train)
ypred2 = clf.predict(X_test_trf2)
accuracy_score(y_test, ypred2)
```

```python
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X, y, test_size=0.2, random_state =2)
```

```python
si2 = SimpleImputer(add_indicator=True)
X_train1 = si2.fit_transform(X_train1)
X_test1 = si2.transform(X_test1)
clf.fit(X_train1, Y_train1)
ypred3 = clf.predict(X_test1)
accuracy_score(Y_test1, ypred3)
```
