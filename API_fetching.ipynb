```python
import requests
import json

all_movies_data = []

headers = {
    "accept": "application/json",
    "Authorization": "Bearer ...."
}


for i in range(1, 531):
    url = f"https://api.themoviedb.org/3/movie/top_rated?language=en-US&page={i}"
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        json_data = json.loads(response.text)
        all_movies_data.extend(json_data['results'])
    else:
        print(f"Error fetching page {i}: {response.status_code}")

print(f"Successfully fetched {len(all_movies_data)} movies.")
```

```python
import pandas as pd
import json

df = pd.DataFrame(all_movies_data)


display(df)
```
