## **Feature Scaling**
Standardization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.sample(5)
```

```python
#adding outliers
df = pd.concat([df, pd.DataFrame({'Age':[3, 70, 80], 'EstimatedSalary':[1000, 300000, 270000], 'Purchased':[1,0,0]})], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.3, random_state=101)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled.head()
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 3))
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title('unscaled')
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], color = 'orange')
ax2.set_title('scaled')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 3))
ax1.set_title('unscaled')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)
ax2.set_title('scaled')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 3))
ax1.set_title('unscaled')
sns.kdeplot(X_train['Age'], ax= ax1)
ax2.set_title('scaled')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 3))
ax1.set_title('unscaled')
sns.kdeplot(X_train['EstimatedSalary'], ax= ax1)
ax2.set_title('scaled')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```
