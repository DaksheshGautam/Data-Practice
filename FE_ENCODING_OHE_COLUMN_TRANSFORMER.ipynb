# Column Transform

```python
from sklearn.model_selection import train_test_split
```

```python
covid = pd.read_csv('covid_toy.csv')
```

```python
covid.head(5)
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(covid.iloc[:,0:5], covid.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
X_train1
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers= [
    ('tnf1', SimpleImputer(), ['fever']),
    ('tnf2', OrdinalEncoder(categories = [['Mild', 'Strong']]), ['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False, drop='first'), ['gender', 'city'])
], remainder='passthrough')

le = LabelEncoder()
y_train1_encoded = le.fit_transform(y_train1)
y_test1_encoded = le.transform(y_test1)
```

```python
X_train1_transformed = transformer.fit_transform(X_train1)
```
